```python
ngbh_gpd = gpd.GeoDataFrame.from_file("./neighbourhoods.geojson")#读取geojson格式文件
ngbh_gpd.head(3)
```

```python
ngbh_gpd.drop("neighbourhood_group",axis=1,inplace=True)
ngbh_gpd.plot()
```

```python
ngbh_gpd.head(3)
```

```python
def df_to_gpd(df,label_longitude,label_latitude):
    df["geometry"]=list(zip(df[label_longitude],df[label_latitude]))
    df["geometry"]=df["geometry"].apply(Point)
    df = gpd.GeoDataFrame(df)
    
#     del df[label_latitude]
#     del df[label_longitude]
    return df

listings_gpd = df_to_gpd(listings_df, "longitude", "latitude")
listings_gpd.info(3)
```

```python
base = ngbh_gpd.plot(color='lightblue', edgecolor='white',figsize=(10, 10))
listings_gpd.plot(ax=base, color='red', marker="o", markersize=50, alpha=0.01) #在底图上叠加餐厅点数据
# plt.gca().xaxis.set_major_locator(plt.NullLocator())#去掉x轴刻度
# plt.gca().yaxis.set_major_locator(plt.NullLocator())#去掉y轴刻度
```

```python
nghd_groupby = listings_gpd.groupby("neighbourhood").size()#根据各区计算总和,groupby返回Series格式
# nghd_groupby = listings_df.neighbourhood.value_counts()#统计dataframe里面的neighbourhood值总和
#上面两个等价
nghd_groupby = nghd_groupby.to_frame().reset_index()#转换为dataframe格式
nghd_groupby.columns = ["neighbourhood","count"]#更改列名
nghd_groupby
#只取出neighbourhood中的汉字并且新建一个地区列
# res1 = []
# for i in range(len(listings_df)):
#     res1.append(''.join(re.findall('[\u4e00-\u9fa5]',listings_df['neighbourhood'].iloc[i]))) #通过Unicode判断中文字符
# listings_df['地区'] = res1
# listings_df.head(5)
# def get_neighbourhood(data):  #对街区数据进行清洗函数,主要用于去除英文，保留中文区域名称
#    if '/' in data:
#        chinese,english = data.split('/')
#        return str(chinese)
#    else:
#        return data
# listings_df['地区'] = listings_df['neighbourhood'].apply(get_neighbourhood)
# listings_df['地区'].head(10)
```

```python
# 左连接合并两个数据表
#SQL select * from ngbh_gpd left join nghd_groupby on ngbh_gpd.neighbourhood = nghd_groupby.neighbourhood
nghd_groupby_join = pd.merge(ngbh_gpd,nghd_groupby,on="neighbourhood",how="left")
nghd_groupby_join
```

```python
# base = nghd_groupby_join.plot(columns="count",cmap='Blues'
#                               ,scheme="boxplot",edgecolor='white',legend=True,figsize=(15,15))
# 画饼状图
lis_dis = listings_df.neighbourhood.value_counts()#统计dataframe里面的neighbourhood值总和 返回为Series
# lis_dis=listings_gpd.groupby("neighbourhood").size()
labels=lis_dis.index
sns.set(font_scale=1.5)
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(15,15))
plt.title('各区listing分布占比',fontdict={'fontsize':18})#标题属性
plt.pie(lis_dis,labels=labels,autopct='%.2f%%',
        explode=[0.1 if i in ['东城区','朝阳区 / Chaoyang','海淀区'] else 0 for i in labels],
        startangle=90,counterclock=False,textprops={'fontsize':12,'color':'black'},
        colors=sns.color_palette('hls',n_colors=18)) 
#东城区，朝阳区，海淀区占比较高，使用explode突出显示
#textprops字体属性
#autopct保留小数点位数
plt.legend(loc='best',shadow=True,fontsize=11)
```
